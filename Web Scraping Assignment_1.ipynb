# 1. Wikepedia.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
Title_names = []
```

```python
titles = soup.find_all('span',class_ ="mw-headline")
titles
```

```python
for i in titles:
    Title_names.append(i.get_text().replace('\n',''))
Title_names
```

```python
import pandas as pd
Title_Names = pd.DataFrame([])
Title_Names['Title_names'] = Title_names
```

```python
Title_Names
```

Final DataFrame of Title_Names of wikipedia Scraping.
